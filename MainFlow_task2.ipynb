```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('01.Data Cleaning and Preprocessing.csv', sep=',')
```

```python
print("\nDataFrame Information:")
df.info()
```

## *Display summary statistics*

```python
print("\nSummary Statistics:")
df.describe()
```

## *Check for missing values in each column*

```python
print("\nMissing alues in Each Column:")
print(df.isnull().sum())
```

## *Handle missing values*

### *Method 1 -  Fill missing values with the mean of their respective columns*

```python
df_filled_mean = df.fillna(df.mean(numeric_only=True))
print("\nDataFrame with missing values filled with mean:")
print(df_filled_mean.head())
```

### *Method 2 - Fill missing values with a specific value (e.g., 0)*

```python
df_filled_zero = df.fillna(0)
print("\nDataFrame with missing values filled with 0:")
print(df_filled_zero.head())
```

### *Method 3 - Drop rows with any missing values*

```python
df_dropped_na = df.dropna()
print("\nDataFrame with rows with any missing values dropped:")
print(df_dropped_na.head())
```

### *Method 4 - Forward fill missing values*

```python

# Forward fill missing values
df_ffill = df.ffill()

# Check for missing values after forward filling
missing_values_after_ffill = df_ffill.isnull().sum()
print("Missing Values in Each Column after forward filling:")
print(missing_values_after_ffill)
```

### *Method 5 - Backward fill missing values*

```python
df_filled_bfill = df.copy()
df_filled_bfill = df_filled_bfill.bfill()
print("\nDataFrame with missing values backward filled:")
print(df_filled_bfill.head())
```

## *Calculate summary statistics*

```python
print("\nSummary Statistics:")
df.describe()
```

## *Calculate correlation matrix*

### *Handling the missing values left after forward filling* 

```python
df_ffill.columns = df_ffill.columns.str.strip()
```

```python
df_ffill['AAWhiteSt-4'] = df_ffill['AAWhiteSt-4'].fillna(df_ffill['AAWhiteSt-4'].mean())
df_ffill['SulphidityL-4'] = df_ffill['SulphidityL-4'].fillna(df_ffill['SulphidityL-4'].mean())
```

```python
missing_values_after_fill = df_ffill.isnull().sum()
print("Missing Values in Each Column after filling:\n", missing_values_after_fill)
```

```python
numeric_df = df_ffill.select_dtypes(include=[np.number])

# Calculate correlation matrix
correlation_matrix = numeric_df.corr()
print("Correlation Matrix:\n", correlation_matrix)
```

## *Plotting the Heatmap for correlation matrix*

```python
plt.figure(figsize=(15, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.show()
```

## **Visualizing the relationships between strong correlations**

### *Pairplot between Upper-HeatT-3 and Lower-HeatT-3*

```python
plt.figure(figsize=(10, 6))
plt.scatter(df_ffill['Upper-HeatT-3'], df_ffill['Lower-HeatT-3'])
plt.xlabel('Upper-HeatT-3')
plt.ylabel('Lower-HeatT-3')
plt.title('Scatter plot of Upper-HeatT-3 vs Lower-HeatT-3')
plt.show()
```

### *Pairplot between Upper-HeatT-3 and AA-Wood-4*

```python
upper_heat_t3 = df_ffill['Upper-HeatT-3']
aa_wood_4 = df_ffill['AA-Wood-4']
plt.figure(figsize=(10, 6))
plt.scatter(upper_heat_t3, aa_wood_4, alpha=0.6)
plt.title('Scatter Plot: Upper-HeatT-3 vs. AA-Wood-4')
plt.xlabel('Upper-HeatT-3')
plt.ylabel('AA-Wood-4')
plt.grid(True)
plt.show()
```
